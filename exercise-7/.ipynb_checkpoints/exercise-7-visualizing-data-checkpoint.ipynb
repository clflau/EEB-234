```python
# loading in the file

with open("/home/eeb177-student/Desktop/eeb-177/class-assignments/classwork-21-Feb/cocoli.txt") as in_file:
    # read in the first line (the header that we don't want)
    in_file.readline()
    # set up empty dictionaries
    sizes_94 = {}
    sizes_97 = {}
    sizes_98 = {}
    
    for line in in_file:
        items = line.split()
        tree_id = items[0]
        
        size_in_94 = float(items[4])
        size_in_97 = float(items[5])
        size_in_98 = float(items[6])
        
        sizes_94[tree_id] = size_in_94
        sizes_97[tree_id] = size_in_97
        sizes_98[tree_id] = size_in_98
        
        
```

```python
import matplotlib.pyplot as plt
# plot some histograms

sizes_in_94 = list(sizes_94.values())
plt.hist(sizes_in_94, log = True)
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharex = True)
ax1.hist(sizes_in_94, log = False)
ax1.set_title('1994 Unlogged')
ax2.hist(sizes_in_94, log = True)
ax2.set_title('1994 logged')
plt.show()
```

```python
sizes_in_94[:] = (value for value in sizes_in_94 if value > 0) # list comprehension to get rid of negative values
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharex = True)
ax1.hist(sizes_in_94, log = False)
ax1.set_title('1994 Unlogged removed negatives')
ax2.hist(sizes_in_94, log = True)
ax2.set_title('1994 logged removed negatives')
plt.show()
```

```python
sizes_in_94 = list(sizes_94.values())
sizes_in_97 = list(sizes_97.values())
# sizes_in_97[:] = (value for value in sizes_in_97 if value > 0)
plt.plot(sizes_in_94, sizes_in_97, 'ro')
plt.xlabel('dbh in 1994')
plt.ylabel('dbh in 1997')
plt.show()
```
